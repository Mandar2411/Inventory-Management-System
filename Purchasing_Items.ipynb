# Inventory Managing System

Purchasing an Item

```python
import json, time
```

```python
items = dict()
```

Function for displaying Products List

```python
def displayProducts():
    fd = open("items.json",'r')
    r = fd.read()
    fd.close()
    
    items = json.loads(r)
    
    for key in items.keys():
        print( "ID : " + key + "  |  Name : " + items[key]['name'] + "  |  Price : " + str(items[key]['price']) + "  |  Quantity Left : " + str(items[key]['quantity']))
        print("\n")
        
    print("\n\n")    
```

Function for purchasing an Item

```python
def purchasingItem(noOfItems):
    bill = 0
    
    # Lists for storing items names and their quantities purchased
    itemsPurchased = []
    quantity = []
    
    fd = open("items.json",'r')
    r = fd.read()
    fd.close()
    
    items = json.loads(r)
    
    for i in range (noOfItems): 
        prod_id = input("Enter product id : ")
        prod_quantity = int(input("Enter the quantity of " + items[prod_id]['name'] + " : "))
        
        # If there are no sufficient items 
        if prod_quantity > items[prod_id]['quantity']:
            prod_quantity = int(input("Enter the quantity less than or equal to " + str(items[prod_id]['quantity']) + " : "))
        
        # Adding amount to bill
        bill += prod_quantity * items[prod_id]['price']
        
        items[prod_id]['quantity'] -= prod_quantity
        
        itemsPurchased.append(items[prod_id]['name'])
        quantity.append(prod_quantity)
        
        
    curr_time = time.ctime()    
    
    fd = open("sales.json",'r')
    r = fd.read()
    fd.close()
    
    sales = json.loads(r)
    
    transaction_id = str(len(sales)+1) 
    
    # Adding item details to Sales Dictionary
    sales[transaction_id] = {"items_list" : itemsPurchased, "quantity_list" : quantity, "billing_amount" : bill, "time" : curr_time}
    
    # Adding entry to sales.json
    js = json.dumps(sales)

    fd = open("sales.json",'w')
    fd.write(js)
    fd.close()   
    
    # Updating items.json (Remaining Quantity of Products)
    js = json.dumps(items)

    fd = open("items.json",'w')
    fd.write(js)
    fd.close()
    
    #Showing Bill
    
    print("\n\n")
    print("You have purchased,")
    
    for i in range (len(itemsPurchased)):
        print(itemsPurchased[i] + " - " + str(quantity[i]))    
    print()    
    print("Total Amount - " + str(bill) + "\n")
    print("Transaction Id : " + transaction_id)
    print("Time : " + curr_time)
    
```

## Run the below cell to purchase an product

```python
noOfItems = int(input("Enter the number of products you want to purchase  : "))
print("\n\n\n")

print("Enter the id of products you want from the below list")
print("\n\n")
displayProducts()

purchasingItem(noOfItems)

print("\nThanks for visiting")
```
